**E-Commerce Sales Analysis**

```python
# Importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the dataset
sales_df = pd.read_csv('Ecommerce_Sales.csv')
```

```python
# Displaying the dataset
sales_df.head(5)
```

```python
# Basic Information
sales_df.info()
```

```python
# Descriptive Summary
sales_df.describe()
```

```python
sales_df.shape
```

```python
sales_df.columns
```

```python
# Checking for the missing values
print(sales_df.isnull().sum())
```

```python
# Filling the missing values
sales_df['Discount'] = sales_df['Discount'].fillna(0)
```

```python
# Finding duplicates
sales_df.duplicated().sum()
```

```python
# Unique customers
print("Unique Customer: ",sales_df['Customer_ID'].nunique())
```

```python
# Transaction per category
print(sales_df['Category'].value_counts().sort_values(ascending=False))

# Visualize the transaction per category
plt.figure(figsize=(12,6))
sns.countplot(x='Category', data=sales_df, palette='rocket')
plt.xticks(rotation=45)
plt.title('Transaction per Category', color='#6A0572')
plt.show()
```

```python
# Total Revenue
print("Total Revenue")
print(sales_df['Total'].sum())
```

```python
# Average Revenue
print("Average Revenue")
print(sales_df['Total'].mean())
```

```python
# Revenue by Category
print("Revenue by Category")
print(sales_df.groupby('Category')['Total'].sum().sort_values(ascending=False) )

# Percentage of revenue by category
print("Percentage of revenue by category")
print(sales_df.groupby('Category')['Total'].sum().sort_values(ascending=False) / sales_df['Total'].sum() * 100)

# Visualize revenue by Category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Total', data=sales_df, palette='rocket')
plt.xticks(rotation=45)
plt.title('Revenue by Category', color='#6A0572')
```

```python
# Total Orders by month only delivered
print("Total Orders by month only delivered")
total_orders = sales_df[sales_df['Order_Status'] =='Delivered'].groupby('Month')['Order_Status'].count()
print(total_orders)
```

```python
# Total Unique Customers
print("Total Unique Customers")
print(sales_df['Customer_ID'].nunique())

# Unique category
print("Unique category")
print(sales_df['Category'].nunique())
```

```python
# Top 10 Customers
print("Top 10 Customers")
print(sales_df.groupby('Customer_ID')['Total'].sum().sort_values(ascending=False).head(10))
```

```python
# convert month into month names
# Extract the month from the 'Date' column and create a new 'Month' column
sales_df['Month'] = pd.to_datetime(sales_df['Date']).dt.month

# Now apply the month name conversion to the new 'Month' column
if pd.api.types.is_numeric_dtype(sales_df['Month']): #check if the Month column is numeric
    sales_df['Month'] = sales_df['Month'].apply(lambda x: datetime.strptime(str(x), '%m').strftime('%B'))
else:
    print("Month column is already in string format (Month names)")

# Revenue by month
print("Revenue by Month")
print(sales_df.groupby('Month')['Total'].sum().sort_values(ascending=False))

# Visualize revenue by MONTH
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Total', data=sales_df, palette='rocket')
plt.title('Revenue by Month', color='#6A0572')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
```

```python
# Revenue by days

# Extract days from date
sales_df['Day'] = pd.to_datetime(sales_df['Date']).dt.day_name()

# Revenue by days
print("Revenue by Days")
print(sales_df.groupby('Day')['Total'].sum().sort_values(ascending=False))

# Visualize the revenue by days
plt.figure(figsize=(12, 6))
sns.barplot(x='Day', y='Total', data=sales_df, palette='rocket')
plt.title('Revenue by Days', color='#6A0572')
```

```python
# Visualizng the revenue share by category in donut chart

plt.figure(figsize=(12, 6))
sales_df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('rocket'))
plt.title('Revenue Share by Category', color='#6A0572')
```

```python
# Total Quantity
print("Total Quantity sold")
print(sales_df['Quantity'].sum())
```

```python
# Customers made repeat purchase
print("Customers made repeat purchase")
print(sales_df[sales_df.duplicated(subset=['Customer_ID'], keep=False)]['Customer_ID'].nunique())
print("")

# Customer who made high purchase
print("Customer who made high purchase")
print(sales_df.groupby('Customer_ID')['Total'].sum().sort_values(ascending=False).head(5))
```

```python
# Average Number of Orders per customer
print("Average Number of orders per customer")
print(sales_df.groupby('Customer_ID')['Order_ID'].count().mean())
```

```python
# Revenue by city
print("Revenue by city")
print(sales_df.groupby('Customer_Location')['Total'].sum().sort_values(ascending=False))

# Visualize revenue by city
plt.figure(figsize=(12,6))
sns.barplot(x='Customer_Location', y='Total', data=sales_df, palette='rocket')
plt.title('Revenue by City', color='#6A0572')
```

```python
# Most Common product by Location
print("Most Common product by Location")
print(sales_df.groupby(['Customer_Location', 'Category'])['Category'].count().sort_values(ascending=False))
```

```python
# Average discount by per category
print("Average discount by per category")
avg=[]
avg = sales_df.groupby('Category')['Discount'].mean().sort_values(ascending=False)
# avg list to datframe
avg = pd.DataFrame(avg)
print(sales_df.groupby('Category')['Discount'].mean().sort_values(ascending=False))

# Visualize in line chart
plt.figure(figsize=(12,6))
sns.lineplot(x='Category', y='Discount', data=avg, color='purple')
plt.title('Average discount by per category', color='#6A0572')
```

```python
# Discount % impact total sales
print("Discount % impact total sales")
print(sales_df.groupby('Discount')['Total'].sum().sort_values(ascending=False))
```

```python
# Do higher discounts lead to higher order volumes?
print("Do higher discounts lead to higher order volumes?")
print(sales_df.groupby('Discount')['Order_ID'].count().sort_values(ascending=False))
```

```python
# Product which sold high without discount
print("Product which sold high without discount")
print(sales_df[sales_df['Discount'] == 0].groupby('Category')['Quantity'].sum().sort_values(ascending=False))
```

```python
# Highest revenue by payment method
print("Highest revenue by payment method")
print(sales_df.groupby('Payment_Method')['Total'].sum().sort_values(ascending=False))
```

```python
#  percentage of Cash on Delivery (COD) orders were canceled
print("percentage of Cash on Delivery (COD) orders were canceled")
print(sales_df[(sales_df['Payment_Method'] == 'Cash on Delivery') & (sales_df['Order_Status'] == 'Canceled')].shape[0]
      / sales_df[sales_df['Payment_Method'] == 'Cash on Delivery'].shape[0] * 100)
```

```python
# Which cities have the highest cancellation rates?
print("Which cities have the highest cancellation rates?")
print(sales_df[sales_df['Order_Status'] == 'Canceled'].groupby('Customer_Location')['Order_Status'].count().sort_values(ascending=False))

cancel_rate = sales_df[sales_df['Order_Status'] == 'Canceled'].groupby('Customer_Location')['Order_Status'].count().sort_values(ascending=False)
cancel_rate = pd.DataFrame(cancel_rate)

# visualize cancellation rate in column chart
plt.figure(figsize=(12,6))
sns.barplot(x='Order_Status', y='Customer_Location', data=cancel_rate, palette='rocket')
plt.title('Cancellation Rate by City', color='#6A0572')
```

```python
# High payment method
print("Most commonly used payment method")
print(sales_df.groupby('Payment_Method')['Payment_Method'].count().sort_values(ascending=False))

# pie chart visualization for high payment method
plt.figure(figsize=(12,6))
sns.countplot(x='Payment_Method', data=sales_df, palette='rocket')
plt.title('Most commonly used payment method', color='#6A0572')
```

```python
# Revenue comparison between Weekdays and Weekends
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekends = ['Saturday', 'Sunday']

# Create a new column 'Day Type'
sales_df['Day_Type'] = sales_df['Day'].apply(lambda x: 'Weekend' if x in weekends else 'Weekday')

# Group revenue by Day Type
revenue_comparison = sales_df.groupby('Day_Type')['Total'].sum()
print('Revenue Comparison between Weekdays and Weekends:')
print(revenue_comparison)

revenue_comparison=pd.DataFrame(revenue_comparison)

# visaulize weekdays vs weekend revenue
plt.figure(figsize=(12,6))
sns.barplot(x='Total', y='Day_Type', data=revenue_comparison, palette='rocket')
plt.title('Revenue comparison between Weekdays and Weekends', color='purple')
```

```python
# Product which has high discount
print("Product which has high discount")
print(sales_df.groupby('Category')['Discount'].sum().sort_values(ascending=False))

# Visualizing the product which has high discount
plt.figure(figsize=(12,6))
sns.barplot(x='Category', y='Discount', data=sales_df, palette='rocket')
plt.title('Product which has high discount', color='#6A0572')
```
